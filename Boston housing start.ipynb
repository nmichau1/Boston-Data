```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor


```

```python
df = pd.read_csv('boston-housing.csv')
df = df.dropna()
df_new = df.drop(['INDUS','AGE'],axis=1)
df_new
df.describe()
```

```python
X = df.drop('MEDV',axis=1)
Y = df['MEDV']


scaler = MinMaxScaler()


X_new = df_new[['RM','LSTAT']]

Y_new = df_new['MEDV']



X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = .2, random_state = 42)
X_new_train,X_new_test,Y_new_train,Y_new_test = train_test_split(X_new,Y_new,test_size = .2, random_state = 42)
```

```python

model = LinearRegression()
model.fit(X_train,Y_train)

model_new = LinearRegression()
model_new.fit(X_new_train, Y_new_train)



results = model.predict(X_test)
print("old MAE: " ,mean_absolute_error(results,Y_test))
print("old R2: ",r2_score(results,Y_test))


results_new = model_new.predict(X_new_test)
print("New MAE: " ,mean_absolute_error(results_new,Y_new_test))
print("New R2: ",r2_score(results_new,Y_new_test))
```

```python
model = RandomForestRegressor()
model.fit(X_train,Y_train)

model_new = LinearRegression()
model_new.fit(X_new_train, Y_new_train)



results = model.predict(X_test)
print("old MAE: " ,mean_absolute_error(results,Y_test))
print("old R2: ",r2_score(results,Y_test))


results_new = model_new.predict(X_new_test)
print("New MAE: " ,mean_absolute_error(results_new,Y_new_test))
print("New R2: ",r2_score(results_new,Y_new_test))
```
